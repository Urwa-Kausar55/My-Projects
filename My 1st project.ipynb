```python
import pandas as pd
df=pd.read_csv('ecommerce_sales_data.csv')
df.dropna() # it removes rows with empty values
```

```python
df.info()#it shows the data type and if there are any missing values
```

```python
df['Total Sales'].mean()# it calculate the average value of data
```

```python
df['Category'].value_counts()# it shows which category has the most orders
```

```python
df['Customer Age'].mean() # it shows which age group does more shopping
```

```python
df.isnull().sum()# it checks if data has any missing values
```

```python
import matplotlib.pyplot as plt
df['Category'].value_counts().plot(kind='bar')
plt.show() # it shows bar grapgh
```

```python
df['Customer Gender'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.show() # it shows ratio of male and female through a pie chart
```

```python
df['Customer Age'].plot(kind='hist' , bins=10, color='orange')
plt.show() #it shows distribution through histogram
```

```python
df['Price'].sum() #it gives the total revenue
```

```python
df.groupby('Category')['Price'].sum().sort_values(ascending=False)# it shows which category has most earning
```
